```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
fact_orders = spark.table("rsa_cata.gold.factorders")
customers   = spark.table("rsa_cata.gold.dimcustomers")
products    = spark.table("rsa_cata.gold.dimproducts")
```

```python
df = (
    fact_orders
    .join(customers, "DimCustomerKey", "left")
    .join(products, col("DimProductKey") == col("product_id"), "left")
)
df.display()
```
